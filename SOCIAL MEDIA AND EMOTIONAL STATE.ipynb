```python
#importing pandas,matplotlib and numpy
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#opening the dataset
dataset=pd.read_excel(r"E:\MYPROJECT\socialstate.xlsx")
```

```python
dataset.info()
```

```python
dataset.head()   #0 means female and 1 means male
```

```python
dataset.head(5)
```

```python
#converting Dominant_Emotion  from string to One Hot Encoding
Emotion=pd.get_dummies(dataset.Dominant_Emotion)
```

```python
Emotion
```

```python
#assigning dependent and independent variables
x=dataset.drop(['Dominant_Emotion','Platform'],axis=1)
y=Emotion
```

```python
x
```

```python
y
```

```python
#now creating train test and split
from sklearn.model_selection import train_test_split
trainx,testx,trainy,testy=train_test_split(x,y,test_size=(0.4))
```

```python
len(trainx),len(testx)
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(trainx,trainy)
```

```python
model.score(testx,testy)
```

```python
import pickle
pickle.dump(model,open('social.pkt','wb'))
```
